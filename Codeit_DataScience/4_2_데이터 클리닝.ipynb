# 06. 데이터 클리닝 : 정확성 2

```python
df = pd.read_csv('data/beer.csv', index_col=0)
df.head()
```

```python
df.plot(kind='box', y='abv')
```

```python
df['abv'].describe()
```

```python
q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3 - q1
```

```python
condition = (df['abv'] < q1 - 1.5*iqr) | (df['abv'] > q3 + 1.5*iqr)
```

```python
df[condition]
```

```python
df.loc[2250, 'abv'] = 0.055
```

```python
condition = (df['abv'] < q1 - 1.5*iqr) | (df['abv'] > q3 + 1.5*iqr)
df[condition].index
```

```python
df.drop(df[condition].index, inplace = True)
```

```python
df.plot(kind='box', y='abv')
```

# 08. 실습과제

```python
df = pd.read_csv('data/movie_metadata.csv')
df.head()
```

```python
q1 = df['budget'].quantile(.25)
q3 = df['budget'].quantile(.75)
iqr = q3-q1
```

```python
df = df[df['budget'] <= q3 + 5*iqr]
df.plot(x = 'budget', y = 'imdb_score', kind = 'scatter')
```
