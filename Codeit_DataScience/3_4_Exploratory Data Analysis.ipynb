# 10. 퀴즈

```python
df = pd.read_csv('data/titanic.csv')
df.head()
```

#### 타이타닉의 승객은 30대와 40대가 가장 많다 - X

```python
df.plot(kind='hist', y='Age')
```

#### 가장 높은 요금을 낸 사람은 30대다 - O

```python
df[df['Fare']==df['Fare'].max()]
```

```python
df.plot(kind='scatter', x = 'Age', y = 'Fare')
```

#### 생존자가 사망자보다 더 많다 - X

```python
df['Survived'].value_counts()
```

#### 1등실, 2등실, 3등실 중 가장 많은 사람이 탑승한 곳은 3등실이다 - O

```python
df['Pclass'].value_counts()
```

#### 가장 생존율이 높은 객실 등급은 1등실이다 - O

```python
p1 = df[df['Pclass']==1]
s_p1 = len(p1[p1['Survived']==1])/len(p1)
s_p1
```

```python
p2 = df[df['Pclass']==2]
s_p2 = len(p2[p2['Survived']==1])/len(p2)
s_p2
```

```python
p3 = df[df['Pclass']==3]
s_p3 = len(p3[p3['Survived']==1])/len(p3)
s_p3
```

```python
sns.kdeplot(df['Pclass'], df['Survived'])
```

#### 나이가 어릴수록 생존율이 높다 - X

```python
df.corr()
```

```python
sns.stripplot(data=df, x = 'Survived', y='Age')
```

```python
sns.violinplot(data=df, x='Survived', y='Age')
```

#### 나이보다 성별이 생존율에 더 많은 영향을 미친다 - O

```python
sns.stripplot(data=df, x='Survived', y='Age', hue='Sex')
```
