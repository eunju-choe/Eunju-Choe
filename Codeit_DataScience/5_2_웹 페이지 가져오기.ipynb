# 08. 실습과제

```python
rating_pages = []
for year in range(2010, 2019):
    for month in range(1, 13):
        for index in range(5):
            url = f'https://workey.codeit.kr/ratings/index?year={year}&month={month}&weekIndex={index}'
            rating_pages.append(requests.get(url).text)
```

```python
print(len(rating_pages))
print(rating_pages[0])
```
